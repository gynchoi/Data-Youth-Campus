```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
import pandas as pd
import time
from tqdm import tqdm
import re
```

# 1. Open the Web Page
Since we get information from the main page, where adult only toons can be shown with mosaic, no login is needed. Get ranking according to day.

```python
driver = webdriver.Chrome('C:/Users/gynchoi/Trap/chromedriver.exe') 
URL = 'https://comic.naver.com/webtoon/weekday.nhn'
driver.get(URL)
```

```python
rankstyle = ['Ntxt_popular','Ntxt_update','Ntxt_count','Ntxt_star_rate']
```

```python
day = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
```

# 2. Popularity

```python
poptitle = []
popday = []
poprank = []

for i in day:
    URL = 'https://comic.naver.com/webtoon/weekdayList?week=' + str(i)
    driver.get(URL)
    time.sleep(0.5)
    
    rankpage = driver.find_element(By.CLASS_NAME, 'Ntxt_popular')
    rankpage.click()
    time.sleep(0.5)
    
    paths = driver.find_elements(By.CLASS_NAME, 'img_list > li > dl > dt > a')
    count = 0
        
    for path in paths:
        title = path.get_attribute('title')
        poptitle.append(title)
        
        ranking = count
        poprank.append(ranking)
        
        popday.append(i)
        
        count += 1
```

```python
# Create and Visualize the dataframe
cols = []
pop = pd.DataFrame(columns=cols)

pop['title'] = poptitle
pop['연재요일'] = popday
pop['인기순'] = poprank
pop
```

# 3. Update
Update information does not affect ranking. This is just sample code.

```python
uptitle = []
uprank = []

for i in day:
    URL = 'https://comic.naver.com/webtoon/weekdayList?week=' + str(i)
    driver.get(URL)
    
    rankpage = driver.find_element(By.CLASS_NAME, 'Ntxt_update')
    rankpage.click()
        
    paths = driver.find_elements(By.CLASS_NAME, 'img_list > li > dl > dt > a')
    count = 0
        
    for path in paths:
        title = path.text
        uptitle.append(title)
        ranking = count
        uprank.append(ranking)
        count += 1
        
up = pd.DataFrame(columns=cols)

up['title'] = uptitle
up['업데이트순'] = uprank
up
```

# 3. Views

```python
cnttitle = []
cntrank = []
cntday = []

for i in day:
    URL = 'https://comic.naver.com/webtoon/weekdayList?week=' + str(i)
    driver.get(URL)
    
    rankpage = driver.find_element(By.CLASS_NAME, 'Ntxt_count')
    rankpage.click()
        
    paths = driver.find_elements(By.CLASS_NAME, 'img_list > li > dl > dt > a')
    count = 0
        
    for path in paths:
        title = path.get_attribute('title')
        cnttitle.append(title)
        
        ranking = count
        cntrank.append(ranking)
        
        cntday.append(str(i))
        count += 1
```

```python
# Create and Visualize the dataframe
cnt = pd.DataFrame(columns=cols)

cnt['title'] = cnttitle
cnt['연재요일'] = cntday
cnt['조회순'] = cntrank

cnt
```

# Star Rating

```python
startitle = []
starday = []
starrank = []

for i in day:
    URL = 'https://comic.naver.com/webtoon/weekdayList?week=' + str(i)
    driver.get(URL)
    
    rankpage = driver.find_element(By.CLASS_NAME, 'Ntxt_star_rate')
    rankpage.click()
        
    paths = driver.find_elements(By.CLASS_NAME, 'img_list > li > dl > dt > a')
    count = 0
        
    for path in paths:
        title = path.get_attribute('title')
        startitle.append(title)
        
        ranking = count
        starrank.append(ranking)
        
        starday.append(str(i))
        
        count += 1
```

```python
# Create and Visualize the dataframe
star = pd.DataFrame(columns=cols)

star['title'] = startitle
star['연재요일'] = starday
star['별점순'] = starrank  
star
```

# 6. Drop Duplications
How to process duplicated ranking?
1. Remove duplicated ranking
2. Convert to mean, median, etc.
3. Rename (e.g '호랑신랑뎐M', '호랑신랑뎐T')

```python
df[df.title == '호랑신랑뎐']
```

```python
titlelist = list(df['title'])
```

```python
title_char = '\/:*?"<>|'
newtilist = []

for title in titlelist:
    for char in title_char:
        if char in title:
            title = title.replace(char,'')
    newtilist.append(title)
    
newtilist 
```

```python
df['title'] = newtilist
```
