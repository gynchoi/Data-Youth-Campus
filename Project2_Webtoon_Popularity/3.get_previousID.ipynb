# Get Previous Work ID
*DATE:* 2022-08-06

*SUMMARY:* Get previous works' IDs if an author who currently updates webtoon has past works. 

*FEATURE:* Use Selenium Webdriver for scraping information. 

**WARNING**\
It is based on the HTML/CSS format from August, 2022. So it may not function correctly now. For correction, check Naver Webtoon homepage's updated HTML/CSS format.

```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import numpy as np
import pandas as pd
import time
from tqdm import tqdm
import re
```

# 1. Open the Web Page

```python
driver = webdriver.Chrome('C:/Users/gynchoi/Trap/chromedriver.exe') 
URL = 'https://comic.naver.com/webtoon/weekday.nhn'
driver.get(URL)
```

```python
infolist = [] 

page = driver.find_elements(By.CLASS_NAME,'title')

for i in tqdm(range(len(page))):
    page = driver.find_elements(By.CLASS_NAME,'title')
    page[i].click()
    time.sleep(0.5)
    
    # If there is only one author - 'class':'other'
    try:
        other = driver.find_element(By.CLASS_NAME, 'other')
        
        # If there is co-workers
        if (driver.find_element(By.CLASS_NAME, 'wrt_nm').text == '웹툰 작가'):   
            print('공동집필 Pass')
            driver.back()
         
        else:    
            other.click()
            time.sleep(0.5)

            info = driver.find_elements(By.CLASS_NAME, 'title > a ')
            idlist = []
            
            for i in info:
                id = re.findall(r'\d+', i.get_attribute('href'))[0]
                idlist.append(id)
            infolist.append(idlist)

            driver.back()
            driver.back()
    
    # If there are co-workers
    except:
        other = driver.find_element(By.CLASS_NAME, 'other2')
        other.click()
        time.sleep(0.5)
        
        # Author1
        path1 = driver.find_element(By.XPATH, r'//*[@id="author_layer_inner"]/li[1]/a')
        path1.click()
        time.sleep(1)
        
        info1 = driver.find_elements(By.CLASS_NAME, 'title > a ')
        idlist = []
        
        for j in info1:
            id = re.findall(r'\d+', j.get_attribute('href'))[0]
            idlist.append(id)
        infolist.append(idlist)
       
        driver.back()
        
        other = driver.find_element(By.CLASS_NAME, 'other2')
        other.click()
        time.sleep(0.5)
        
        # Author2
        path2 = driver.find_element(By.XPATH, r'//*[@id="author_layer_inner"]/li[2]/a')
        path2.click()
        time.sleep(0.5)
        
        info2 = driver.find_elements(By.CLASS_NAME, 'title > a ')
        idlist = []
        
        for k in info2:
            id = re.findall(r'\d+', k.get_attribute('href'))[0]
            idlist.append(id)
        infolist.append(idlist)
    
        driver.back()
         
        # Check if there is third author
        other = driver.find_element(By.CLASS_NAME, 'other2')
        other.click()
        time.sleep(0.5)
        
        try:
            path3 = driver.find_element(By.XPATH, r'//*[@id="author_layer_inner"]/li[3]/a')
            path3.click()
            time.sleep(0.5)

            info3 = driver.find_elements(By.CLASS_NAME, 'title > a ')
            idlist = []
            
            for l in info3:
                id = re.findall(r'\d+', l.get_attribute('href'))[0]
                idlist.append(id)
            infolist.append(idlist)
            
            driver.back()
            driver.back()
            
        except:
            driver.back()
```

```python
# Create and Visualize the dataframe
df = pd.DataFrame(infolist)
df
```

```python
# Save as CSV format
df.to_csv('최종_네이버웹툰_과거작품정보(1).csv', encoding = 'utf-8-sig', index=None)
```
